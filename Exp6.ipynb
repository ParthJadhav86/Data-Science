```python
import pandas as pd

# Load the dataset
file_path = '/content/Cost_of_Living_Index_by_Country_2024.csv'  # Use the correct path if different
df = pd.read_csv(file_path)

# Display the first few rows to confirm loading
df.head()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter Plot: Cost of Living Index vs Rent Index
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Cost of Living Index', y='Rent Index', hue='Country')
plt.title('Scatter Plot: Cost of Living Index vs Rent Index')
plt.xlabel('Cost of Living Index')
plt.ylabel('Rent Index')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Bar Plot: Top 10 Countries by Cost of Living Index
top_10_countries = df.nlargest(10, 'Cost of Living Index')
plt.figure(figsize=(12, 6))
sns.barplot(data=top_10_countries, x='Country', y='Cost of Living Index')
plt.title('Bar Plot: Top 10 Countries by Cost of Living Index')
plt.xlabel('Country')
plt.ylabel('Cost of Living Index')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram: Distribution of Cost of Living Index
plt.figure(figsize=(10, 6))
sns.histplot(df['Cost of Living Index'], kde=True, bins=20)
plt.title('Histogram: Distribution of Cost of Living Index')
plt.xlabel('Cost of Living Index')
plt.ylabel('Frequency')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Line Chart: Cost of Living Index for the Top 10 Countries
plt.figure(figsize=(12, 6))
sns.lineplot(data=top_10_countries, x='Country', y='Cost of Living Index', marker='o')
plt.title('Line Chart: Cost of Living Index for the Top 10 Countries')
plt.xlabel('Country')
plt.ylabel('Cost of Living Index')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14, 8))
sns.boxplot(data=df[['Cost of Living Index', 'Rent Index',
                     'Cost of Living Plus Rent Index', 'Groceries Index',
                     'Restaurant Price Index', 'Local Purchasing Power Index']])
plt.title('Box Plot: Distribution of Various Indexes')
plt.xlabel('Index Type')
plt.ylabel('Value')
plt.show()
```
